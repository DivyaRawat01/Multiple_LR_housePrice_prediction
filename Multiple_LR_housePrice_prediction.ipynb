```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset=pd.read_csv("AmesHousing_simple.csv")
```

```python
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
dataset["BsmtFin SF 1"]=dataset["BsmtFin SF 1"].fillna(dataset["BsmtFin SF 1"].mode()[0])
```

```python
dataset["Bsmt Unf SF"]=dataset["Bsmt Unf SF"].fillna(dataset["Bsmt Unf SF"].mode()[0])
dataset["Total Bsmt SF"]=dataset["Total Bsmt SF"].fillna(dataset["Total Bsmt SF"].mode()[0])
```

```python
input_x=dataset.iloc[:,:-1]
output_y=dataset["SalePrice"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(input_x,output_y,test_size=0.2,random_state=19)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
arr=lr.predict(x_test)
```
